# Executive Summary

The purpose of this case study is to predict the  total of points for an NHL hockey player based on the features: weight, height, hall of fame, position and body mass index.
All data comes from the official NHL page.

\
During the data cleaning process, columns that did not influence the prediction were deleted: nationality, state, province, draft round, country, year of draft, overall, date of birth, city of birth, first season and player name.

In addition, the variables were converted: weight, height, number of points and number of assistances from object to integer to generate the `points_per_game` variable and calculate the total points divided by the total number of games played.

To facilitate the classification, a range of points based on points per game was created, from a to e, being `range_a` best range and `range_e` worst performance

\
Graphs were plotted for all players comparing BMI and total points in addition to comparison graphs by position, where we can see weight, height, bmi, hall of fame by position.

\

We trained Decision Trees models using the NHL dataset using differents hyperparameters to find the best model. 
After evaluating all the models using metrics such:  accuracy, F1 score, recall, and precision metrics, we found that the best model was found to be the one with a maximum depth=4, min samples leaf = 2 and using the entropy.

We have columns ( weight, height, assist, points) that should have as value , int64 ,but they are as object. \

for all the object columns we change to string

```python
# convert points to a integer
df['points'] = df['points'].str.replace(',', '').astype(float)
df['points'] = df['points'].astype(int)
df.head()
```

```python
# convert assist to a integer

df['assist'] = df['assist'].str.replace(',', '').astype(float)
df['assist']  = df['assist'].astype(int)
df.head()
```

```python
# convert games_played to a integer

df['games_played'] = df['games_played'].str.replace(',', '').astype(float)
df['games_played'] = df['games_played'].astype(int)
df.head()
```

```python
# convert weight to a integer and remove rows with non int values
df.drop(df[df.weight == '--'].index, inplace=True)


df['weight'] = df['weight'].str.replace(',', '').astype(float)
df['weight'] =df['weight'].astype(int)
df.head()
```

```python
# convert height to a integer and remove rows with non int values
df.drop(df[df.height == '--'].index, inplace=True)

df['height'] = df['height'].str.replace(',', '').astype(float)
df['height'] = df['height'].astype(int)
df.head()
```

```python
# convert object to string for easy operations

df['position'] = df['position'].astype('string')
df['hall_fame'] = df['hall_fame'].astype('string')

df.head()
```

```python
df.info()
```

We need to relate the total of goals, assists and points by the amount of games played

```python
df.head()
```

```python
# new variable points_per_game, total of points divided by games played

df['points_per_game'] = df['points'] / df['games_played']
df.head() 
```

```python
# create new variable bmi (body mass index), relation between weight and height

df['bmi'] = (df['weight']/(df['height']**2)* 703)
df.tail()
```

###Creating range for the points

* range_e -->  0 - 0.40
* range_d -->  0.40 - 0.80
* range_c -->  0.80 - 1.2
* range_b -->  1.20 - 1.6
* range_a -->  1.60 - 2.0

```python
min_points = df['points_per_game'].min()
max_points = df['points_per_game'].max()

print(min_points)
print(max_points)
```

```python
# range_e = [0,0.40]
# range_d = [0.41, 0.80]
# range_c = [0.81, 1.2]
# range_b =[1.21, 1.6]
# range_a = [1.61, 2.0]

def label_ranges (row):
  if row['points_per_game']>=0.0 and row['points_per_game']<=0.40:
      return'range_e'
    
  if row['points_per_game']>0.4 and row['points_per_game']<=0.80:
      return 'range_d'

  if row['points_per_game']>0.8 and row['points_per_game']<=1.20:
      return 'range_c'

  if row['points_per_game']>1.20 and row['points_per_game']<=1.60:
      return 'range_b'

  if row['points_per_game']>1.60 and row['points_per_game']<=2.0:
      return 'range_a'

df['range_points'] = df.apply(lambda row: label_ranges(row), axis=1)
df['range_points'] = df['range_points'].astype('string')

df.head()
```

## Model

Explain what the following hyperparameters control: 

- max_depth: This parameter defines the maximum depth of the decision tree. The deeper you can get the better fit in training you may be at risk of overfitting.

- max_features: This parameter defines the maximum number of features that are considered for a node.

- min_impurity_decrease: This parameter defines a limit for the division of nodes.

- min_samples_leaf: This parameter defines the minimum number of samples needed to be in a leaf node. If the number of samples in a node is less than this value, the node is no longer split.

- min_samples_split: This parameter defines the minimum number of samples needed to split an internal node. If the number of samples in an internal node is less than this value, the node is no longer split.

Based on the table with results, the best performance appears in the:
- model: Entropy
- Max Depth: 4
- number of samples required to be a leaf node : 1

The validation accuracy is 0.99 and the test accuracy is 0.993, also the other values are similar, some other models show a difference in the precision values between validation and testing, this may indicate an overfitting since the model has great metrics, but when tested with the test set, data that it does not know, the metrics drop.

```python
# Define the hyperparameters to try
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2, 4, 6],  # max depth for the decision tree
    'min_samples_leaf': [2,4],  # minimum number of samples required to be at a leaf node
}

# Create a table to display the results
table = PrettyTable(['Hyperparam', 'Max Depth', 'Min Samples Leaf', 
                     'Validation Accuracy',
                     'Validation Precision',
                     'Validation Recall', 
                     'Validation F1 Score', 
                     ])

def combine_models(params):
    # Combine hyperparameters and run a decision tree for each option
    for c in params['criterion']:
        for max_depth in params['max_depth']:
            for min_samples_leaf in params['min_samples_leaf']:
                dtc = DecisionTreeClassifier(criterion=c, max_depth=max_depth, min_samples_leaf=min_samples_leaf)
                # Train the model with the training set
                dtc.fit(X_train, y_train)
                # Make predictions on the validation 
                y_val_pred = dtc.predict(X_val)

                # Calculate the performance metrics
                val_acc, val_prec, val_rec, val_f1 = calculate_metrics(y_val, y_val_pred)

                # Add the results to the table
                table.add_row([c, max_depth, min_samples_leaf, 
                               val_acc, 
                               val_prec,
                               val_rec,
                               val_f1,
                               ])
                # call function to show the confusion matrix
                title_cm = ((f'(criterion={c}, max_depth={max_depth}, min_samples_leaf={min_samples_leaf})'))
                display_confusion_matrix(y_val, y_val_pred, title_cm)

    print(table)

def calculate_metrics(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='macro')
    rec = recall_score(y_true, y_pred, average='macro')
    f1 = f1_score(y_true, y_pred, average='macro')
    return acc, prec, rec, f1

def display_confusion_matrix(y_true, y_pred, title_cm):
    # Compute the confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Plot the confusion matrix
    # sorted the lables to put in order
    labels = sorted(list(set(y_true)))
    sns.heatmap(cm, annot=True, cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('True Values')
    print("=" * 80)
    print(title_cm)
    print("=" * 80)
    plt.show()

combine_models(params)
```

## Define the best model
